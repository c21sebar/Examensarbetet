```python
import random
def generateNewData():
  numberOfFiles = 10
  for index in range(numberOfFiles):
    data1 = [93, 98, 133, 128, 95, 98, 131, 118, 108, 122]
    data2 = [107, 342, 328, 118, 477, 270, 283, 176, 399, 97]
    data3 = [167, 199, 261, 135, 113, 246, 159, 225, 157, 229]    
    for i,val in enumerate(data1):
      add_number = random.randint(1,10)
      data1[i] = data1[i] + add_number     
    for i,val in enumerate(data2):
      add_number = random.randint(1,10)
      data2[i] = data2[i] + add_number     
    for i,val in enumerate(data3):
      add_number = random.randint(1,10)
      data3[i] = data3[i] + add_number    
    print(f' {index} Data1: {data1}, Data2: {data2}, Data3: {data3}')
    df = pd.DataFrame({'Chrome': data1, 'IE': data2, 'Firefox': data3})
   
    file_name = "data" + str(index)+".txt"        
    df.to_csv(file_name, index=False, header=False)  
generateNewData()
    
```

```python

import matplotlib.pyplot as plt
def exampleBarChars():
  #Read your data from file
  file = "mergedfile.txt"
  df = pd.read_csv(file, sep=",", header=None, names=['Chrome','Explorer','Firefox'])
  # width of the bars
  barWidth = 0.6
  # Bars Data
  barsData = df.mean()
   # The x-position order of bars
  barsOrder = range(len(df.columns))
  # Std Bars Interval
  #barsInterval = df.std()
  # Standard Error Bars intervals
  barsInterval = df.sem()
  # Colours of bar charts
  colors=["red", "green", "blue"]
  # Opacity of colours
  Opacity=0.5
  # Start values from bottom of the bars
  minValue=df.values.min()
  # Interval cap size
  intervalCapsize=7
  # Plot bars
  plt.bar(barsOrder, barsData, color = colors , edgecolor = 'black', width = barWidth, yerr=barsInterval, capsize=7, alpha=Opacity, bottom=intervalCapsize)
  #Put a tick on the x-axis undex each bar and label it with column name
  plt.xticks(range(len(df.columns)), df.columns)
  plt.ylabel('Load time')
  #plt.xlabel('Browsers')
  plt.title('Load-Time Means Comparison')
  plt.show()
  plt.savefig("S.E_BarChars.png")
exampleBarChars()
```

```python
def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m, se = np.mean(a), stats.sem(a)
  h = se * stats.t.ppf((1 + confidence) / 2., n-1)
  return -h, +h

#Read your data from file
file = "mergedfile.txt"
df = pd.read_csv(file, sep=",", header=None, names=['Chrome','Explorer','Firefox'])
#Call CI function
print ("CIs Chrome: ", mean_confidence_interval(df['Chrome']))
print ("CIs Explorer", mean_confidence_interval(df['Explorer']))
print ("CIs Firefox:", mean_confidence_interval(df['Firefox'])) 

CI_Chrome = mean_confidence_interval(df['Chrome'])
CI_Explorer = mean_confidence_interval(df['Explorer'])
CI_Firefox = mean_confidence_interval(df['Firefox'])


# width of the bars
barWidth = 0.6

# Bars Data
barsData = df.mean()

# The x-position order of bars
barsOrder = range(len(df.columns))

# Std Bars Interval
#barsInterval = df.std()

# Bars for CI Intervals
df_CI = pd.DataFrame(list(zip(CI_Chrome, CI_Explorer,CI_Firefox)), columns = ['Chrome', 'Explorer', 'Firefox'])
barsInterval = df_CI.iloc[1]

# Colours of bar charts
colors=["red", "green", "blue"]

# Opacity of colours
Opacity=0.5

# Start values from bottom of the bars
minValue=df.values.min()

# Interval cap size
intervalCapsize=7

# Plot bars
plt.bar(barsOrder, barsData, color = colors , edgecolor = 'black', width = barWidth, yerr=barsInterval, capsize=7, alpha=Opacity, bottom=intervalCapsize)

#Put a tick on the x-axis undex each bar and label it with column name
plt.xticks(range(len(df.columns)), df.columns)

plt.ylabel('Load time')

#plt.xlabel('Browsers')
plt.title('Load-Time Means Comparison')
plt.show()
```
